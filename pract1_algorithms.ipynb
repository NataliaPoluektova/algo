### Завдання 2
**Приклади програм знаходження суми n чисел**

1. Через цикл
2. Через формулу


```
# 1. Сума через цикл
def sum_cycle(n):
    total = 0
    for i in range(1, n+1):
        total += i
    return total

print("Сума через цикл (n=10):", sum_cycle(10))

# 2. Сума через формулу
def sum_formula(n):
    return n * (n + 1) // 2

print("Сума через формулу (n=10):", sum_formula(10))
```

### Завдання 3
**Приклад алгоритму – Бінарний пошук**

```
# Бінарний пошук (приклад)
def binary_search(lst, item):
    low = 0
    high = len(lst) - 1

    while low <= high:
        mid = (low + high) // 2
        guess = lst[mid]
        if guess == item:
            return mid
        if guess > item:
            high = mid - 1
        else:
            low = mid + 1
    return None

my_list = [1, 3, 5, 7, 9]
print("Пошук 3:", binary_search(my_list, 3))
print("Пошук 10:", binary_search(my_list, 10))
```
