def selection_sort(array):
    length = len(array)
    # Перебираємо кожен елемент масиву (крім останнього)
    for start_index in range(length - 1):
        # Індекс найменшого елемента у поточній ітерації
        smallest_index = start_index

        # Шукаємо менший елемент у решті масиву
        for current_index in range(start_index + 1, length):
            if array[current_index] < array[smallest_index]:
                smallest_index = current_index

        # Міняємо місцями знайдений найменший з поточним
        array[start_index], array[smallest_index] = array[smallest_index], array[start_index]

def bubble_sort(a):
    size = len(a)
    for i in range(size):
        for j in range(size - 1, i, -1):
            if a[j - 1] > a[j]:  # якщо попередній більший за поточний
                a[j - 1], a[j] = a[j], a[j - 1]
