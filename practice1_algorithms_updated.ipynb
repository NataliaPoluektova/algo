## 2. Оцінювання алгоритмів

Алгоритми можна порівнювати за:
- правильністю;
- зрозумілістю;
- кількістю кроків;
- швидкодією;
- використанням пам'яті.



### Приклад: сума чисел від 1 до n

**Варіант 1 (через цикл):**


````
# Python
n = 1000000
s = 0
for i in range(1, n+1):
    s += i
print(s)
````

````
// C++
#include <iostream>
using namespace std;
int main() {
    long long n = 1000000, s = 0;
    for (long long i = 1; i <= n; i++) s += i;
    cout << s;
    return 0;
}
````


**Варіант 2 (через формулу):**


````
# Python
n = 1000000
s = n * (n + 1) // 2
print(s)
````

````
// C++
#include <iostream>
using namespace std;
int main() {
    long long n = 1000000;
    long long s = n * (n + 1) / 2;
    cout << s;
    return 0;
}
````

## 3. Перший алгоритм

Задача: знайти найбільше число у списку.


````
# Python
arr = [5, 9, 2, 8, 1, 7]
max_val = arr[0]
for x in arr:
    if x > max_val:
        max_val = x
print("Максимальне число:", max_val)
````

````
// C++
#include <iostream>
#include <vector>
using namespace std;
int main() {
    vector<int> arr = {5, 9, 2, 8, 1, 7};
    int max_val = arr[0];
    for (int x : arr) if (x > max_val) max_val = x;
    cout << "Максимальне число: " << max_val;
    return 0;
}
````


Тепер порівняйте зі стандартним алгоритмом Евкліда:


````
# Python
def gcd(a, b):
    while b != 0:
        a, b = b, a % b
    return a

print(gcd(48, 18))
````

````
// C++
#include <iostream>
using namespace std;
int gcd(int a, int b) {
    while (b != 0) {
        int t = b;
        b = a % b;
        a = t;
    }
    return a;
}
int main() {
    cout << gcd(48, 18);
    return 0;
}
````

## 5. «O-велике» (оцінка складності)

### Приклади:
- Лінійний пошук у списку: **O(n)**
- Подвійний цикл (порівняння всіх елементів): **O(n²)**
- Бінарний пошук: **O(log n)**
- Алгоритм Евкліда: **O(log min(a, b))**
